```python
import os
from dotenv import load_dotenv

# .env 파일에서 환경변수 로드
load_dotenv()

# API 키 가져오기
API_KEY = os.getenv("DATA_GO_KR_API_KEY")

# 잘 불러왔는지 확인 (앞 10글자만 출력 — 전체 노출 방지)
print(API_KEY[:10])
```

```python
import requests

# 목록조회 엔드포인트 URL
url = "http://apis.data.go.kr/B554287/NationalWelfareInformationsV001/NationalWelfarelistV001"

# 필수 파라미터
params = {
    "serviceKey": API_KEY,
    "callTp": "L",  # L: 목록조회
    "pageNo": 1,  # 1페이지부터
    "numOfRows": 10,  # 10건 가져오기
    "srchKeyCode": "003",  # 제목+내용 검색
}

# GET 요청 보내기
response = requests.get(url, params=params)

# 응답 상태 확인
print("상태 코드:", response.status_code)
print("응답 내용 (앞 500자):")
print(response.text[:500])
```

```python
import xml.etree.ElementTree as ET

# XML 파싱
root = ET.fromstring(response.text)

# 전체 건수 확인
total_count = root.find("totalCount").text
print(f"전체 복지서비스 수: {total_count}건")

# 첫 번째 서비스 데이터의 모든 필드 확인
first_item = root.find("servList")
if first_item is not None:
    print("\n--- 첫 번째 서비스 데이터 필드 ---")
    for child in first_item:
        print(f"{child.tag}: {child.text}")
```

```python
# 옵션 파라미터 추가
params_filtered = {
    "serviceKey": API_KEY,
    "callTp": "L",
    "pageNo": 1,
    "numOfRows": 10,
    "srchKeyCode": "003",
    "lifeArray": "006",  # 생애주기: 청년
}

response_filtered = requests.get(url, params=params_filtered)
root_filtered = ET.fromstring(response_filtered.text)

total = root_filtered.find("totalCount").text
print(f"청년 대상 복지서비스: {total}건")

# 서비스명만 출력
for item in root_filtered.findall("servList"):
    name = item.find("servNm").text
    dept = item.find("jurMnofNm").text
    print(f"- {name} ({dept})")
```
